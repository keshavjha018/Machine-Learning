**1.2 Import Dataset**

```python
import pandas as pd
path = "/content/drive/MyDrive/Colab Notebooks/Project1/music.csv"
music_data = pd.read_csv(path)
# print(music_data)

# Summary/overview of dataset -
# df.describe()
```

**2. Cleaning The data**
 - Divide into input & output

```python
# Input Data
X = music_data.drop(columns=['genre'])

# Output Data
Y = music_data['genre']
```

**3. Learning/Training & Predictions**
- Based on Input Data & Expected Output we'll train the model

We are using an algo called Decision Tree. - Just Import & use

Then we ll predict the output for the new input which we havn't feeded our model

```python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier()
model.fit(X, Y)

# Predict - 21 yr old male, 22 yr old Female
predictions = model.predict([ [21, 1], [22,0] ])
print(predictions)
```

**4. Calculating Accuracy**
- Divide the dataset into 80:20 ratio.
80% of data will be used for training
& 20% will be used for testing

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3)

# Train Model
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)

# Predict Outputs of test data
predictions = model.predict(X_test)

# Compare predictions with actual output
score = accuracy_score(Y_test, predictions)
print("Accuracy: ",score*100, "%")
```
